```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import cv2
from PIL import Image
from keras.layers import Dense, Dropout, Conv2D, MaxPool2D, Flatten
from keras import layers,models
```

```python
train_path='../input/gender-recognition-200k-images-celeba/Dataset/Train'
test_path='../input/gender-recognition-200k-images-celeba/Dataset/Test'

male_train='../input/gender-recognition-200k-images-celeba/Dataset/Train/Male'
female_train='../input/gender-recognition-200k-images-celeba/Dataset/Train/Female'

male_test='../input/gender-recognition-200k-images-celeba/Dataset/Test/Male'
female_test='../input/gender-recognition-200k-images-celeba/Dataset/Test/Female'
```

```python
category=os.listdir(train_path)
num_category=len(category)
print(num_category)
```

```python
Train_images=[]
for cate in category:
     folder = train_path +"/"+ cate
     Train_images.append(len(os.listdir(folder)))
```

```python
Test_images=[]
for cate in category:
     folder = test_path +"/"+ cate
     Test_images.append(len(os.listdir(folder)))
```

```python
fem_train=[]
label1=[]
for i in os.listdir(female_train):
    if os.path.isfile(female_train + '/' + i):
        female=Image.open(female_train + '/' + i)
        female=female.resize((50,50),Image.ANTIALIAS).convert("L")
        female=np.asarray(female)/255.0
        
        fem_train.append(female)
        label1.append(0)
```

```python
mal_train=[]
for i in os.listdir(male_train):
    if os.path.isfile(male_train + '/' + i):
        male=Image.open(male_train + '/' + i)
        male=male.resize((50,50),Image.ANTIALIAS).convert("L")
        male=np.asarray(male)/255.0
        
        mal_train.append(male)
        label1.append(1)
```

```python
x_train=np.concatenate((fem_train,mal_train),axis=0)
y_train=np.asarray(label1)
y_train=y_train.reshape(x_train.shape[0],1)
```

```python
fem_test=[]
label2=[]
for i in os.listdir(female_test):
    if os.path.isfile(female_test + '/' + i):
        female2=Image.open(female_test + '/' + i)
        female2=female2.resize((50,50),Image.ANTIALIAS).convert("L")
        female2=np.asarray(female2)/255.0
        
        fem_test.append(female2)
        label2.append(0)
```

```python
mal_test=[]
for i in os.listdir(male_test):
    if os.path.isfile(male_test + '/' + i):
        male2=Image.open(male_test + '/' + i)
        male2=male2.resize((50,50),Image.ANTIALIAS).convert("L")
        male2=np.asarray(male2)/255.0
        
        mal_test.append(male2)
        label2.append(1)
```

```python
x_test=np.concatenate((fem_test,mal_test),axis=0)
y_test=np.asarray(label2)
y_test=y_test.reshape(x_test.shape[0],1)
```

```python
x=np.concatenate((x_train,x_test),axis=0)
y=np.concatenate((y_train,y_test),axis=0)
x=x.reshape(x.shape[0],x.shape[1]*x.shape[2])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,random_state=2,test_size=0.2)
```

```python
xtrain = X_train.T
xtest = X_test.T
ytrain = Y_train.T
ytest = Y_test.T
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_n = KNeighborsClassifier(5).fit(X_train,Y_train)
```

```python
predKNN=knn_n.predict(X_test)
```

```python
#calculating precision and reall
precisionKNN = precision_score(Y_test,predKNN)
recallKNN = recall_score(Y_test,predKNN)
```

```python
print('Precision KNN: ',precisionKNN)
print('Recall KNN: ',recallKNN)
```

```python
#report KNN
print(classification_report(Y_test,predKNN))
```

```python
print("Accuracy Score Train KNN: %{:.2f} ".format(accuracy_score(knn_n.predict(X_train),Y_train)*100))
print("Accuracy Score Test KNN: %{:.2f} ".format(accuracy_score(knn_n.predict(X_test),Y_test)*100))
```
